## 1
* $\min_{||x-c||_2 \leq r} ||x - z||_2$ is equivalent to $\min_{||x-c||_2^2 \leq r^2} ||x - z||_2^2$
* Lagrangian $L(x, \lambda) = ||x - z||_2^2 + \lambda (||x-c||_2^2 - r^2), \ \lambda \ge 0$
* $\nabla_x L(x, \lambda) = 2(x - z) + 2 \lambda (x - c) \rightarrow x^* = \frac{1}{1+\lambda} (z + \lambda c)$
* $\Delta_x L(x, \lambda) = (2 + 2 \lambda) I \succ 0 \rightarrow \min $
* Dual function $d(\lambda) = \inf_x L(x, \lambda) = \frac{\lambda^2}{(1 + \lambda)^2} ||z - c||_2^2 + \frac{\lambda}{(1 + \lambda)^2} ||z - c||_2^2 - \lambda r^2 =  \frac{\lambda}{(1 + \lambda)} ||z - c||_2^2 - \lambda r^2$
* $\frac{\partial d(\lambda)}{\partial \lambda} = \frac{1}{(1 + \lambda)^2} ||z - c||_2^2 - r^2 \rightarrow \lambda^* = \frac{||z - c||}{r} - 1$
* $\frac{\partial^2 d(\lambda)}{\partial \lambda^2} = \frac{-2}{(1 + \lambda)^3} ||z - c||_2^2 \leq 0 \rightarrow \max$
* Answer: $x_{min} = \frac{r}{||z - c||} (z - c + \frac{||z - c||}{r}c) = \frac{z-c}{||z-c||}r + c$

## 2
* $\min_{(a, x) = b} ||x - z||_2$ is equivalent to $\min_{(a,x)=b} ||x - z||_2^2$
* Lagrangian $L(x, \nu) = ||x - z||_2^2 + \nu ((a,x) - b), \ \lambda \ge 0$
* $\nabla_x L(x, \nu) = 2(x - z) + \nu a \rightarrow x^* = \frac{2z - \nu a}{2}$
* $\Delta_x L(x, \nu) = 2 I \succ 0 \rightarrow \min $
* Dual function $d(\nu) = \inf_x L(x, \nu) = \nu ((z, a) - b) - \nu^2 \frac{||a||^2}{4}$
* $\frac{\partial d(\nu)}{\partial \nu} = ((z, a) - b) - \nu \frac{||a||^2}{2}\rightarrow \nu^* = 2\frac{(z, a) - b}{(a, a)}$
* $\frac{\partial^2 d(\nu)}{\partial \nu^2} = - \frac{||a||^2}{2} \leq 0 \rightarrow \max$
* Answer: $x_{min} = z - \frac{(z, a) - b}{(a, a)} a$

## 3
* Lagrangian $L(x, \nu) = f(x) + (\nu, Ax - b), \ \nu \in \mathbb R^m$
* $d(\nu) = \inf_x (f(x) + (\nu, Ax - b)) = -\sup_x (-(\nu, Ax-b) - f(x)) = -\sup_x((-A^T \nu, x) - f(x)) - (\nu, b)$
* $d(\nu) = -f^*(-A^T \nu) - (\nu, b)$

## 4
* Lagrangian $L(x,\lambda)=(c,x)+\lambda ((Ax,x)-1), \ \lambda \ge 0$
* $\nabla_x L(x,\lambda)=c+\lambda(A + A^T)x= \ [A = A^T] \ =c + 2\lambda Ax \rightarrow x^* = -\frac{1}{2\lambda}A^{-1}c$
* $\Delta_x L(x, \lambda) = 2\lambda A \succ 0 \rightarrow \min$
* $d(\lambda) = -\frac{1}{4\lambda}(c, A^{-1}c) - \lambda$
* Dual probem: $\max_{\lambda} (-\frac{1}{4\lambda}(c, A^{-1}c) - \lambda)$
* $\frac{\partial d(\lambda)}{\partial \lambda} = \frac{1}{4\lambda^2} (A^{-1}c, c) - 1 \rightarrow \lambda^*=\sqrt{\frac{(A^{-1}c, c)}{4}}$
* $\frac{\partial^2 d(\lambda)}{\partial \lambda^2} = -\frac{1}{2\lambda^3} (A^{-1}c, c) \leq 0 \rightarrow \max$
* Answer: $x_{min} = -\frac{1}{2\sqrt{(A^{-1}c, c)}}A^{-1}c$

## 5
* Feasible set: $2 \leq x \leq 4$.
* Minimum of $f$ at $x=0$ is infeasible, so the optimal solution is either $x=2$ or $x=4$. $f(2) \leq f(4)$, so $x^*=2$.
* Optimal value $p^* = f(2) = 5$
* Lagrangian $L(x, \lambda) = x^2 + 1 + \lambda (x-2)(x-4), \ \lambda \ge 0$
* $\frac{\partial L(x, \lambda)}{\partial x} = 2x + 2x\lambda - 6\lambda \rightarrow x^* = \frac{3 \lambda}{\lambda + 1}$
* $\frac{\partial^2 L(x, \lambda)}{\partial x^2} = 2\lambda \ge 0 \rightarrow \min$
* $d(\lambda) = \frac{-\lambda^3 + 8 \lambda^2 + 10 \lambda + 1}{(\lambda + 1)^2} = \frac{- \lambda^2 + 9\lambda + 1}{\lambda + 1}$
* $\frac{\partial d(\lambda)}{\partial \lambda} = - \frac{\lambda^2 + 2\lambda - 8}{(\lambda + 1)^2} \rightarrow \lambda^* = 2$ (positive root of the numerator)
* $\frac{\partial^2 d(\lambda)}{\partial \lambda^2} = -\frac{18}{(\lambda + 1)^3} \leq 0 \rightarrow \max$
* The maximization problem is concave since the second derivative is always negative for $\lambda \le 0$
* Answer: $x_{min}=2, \ p^* = d(2) = 5$ which coincides (strong duality holds) with the answer we got solving the primal problem.

```python
%pylab inline

# plots of the objective, feasible set, lagrangian

x = np.linspace(-1, 5, 500)
y = np.power(x, 2) + 1
l1 = np.power(x, 2) + 1 + 0.5 * (x - 2) * (x - 4)
l2 = np.power(x, 2) + 1 + 1 * (x - 2) * (x - 4)
l3 = np.power(x, 2) + 1 + 2 * (x - 2) * (x - 4)

plt.figure(figsize=(10, 10))
plt.plot(x, y, label='f(x)', c='red')
plt.plot(2, 5, marker='*', markersize=15, color='red', label='optimal point')
plt.plot(x, l1, label='lambda = 0.5')
plt.plot(x, l2, label='lambda = 1')
plt.plot(x, l3, label='lambda = 2')
plt.axvline(x=2, label='left border', c='black')
plt.axvline(x=4, label='right border', c='black')
plt.legend()
```

## 6
* There must be a type in the problem description: $Ax \le b$ does not make any sense, so henceforth I assume that the constraint is $Ax = b$
* Lagrangian: $L(x, \nu) = \frac{1}{2}x^TPx+(x,q)+(\nu, Ax-b)$
* $\nabla_x L(x,\nu)=Px + q + A^T\nu \rightarrow x^* = -P^{-1}(A^T\nu + q)$
* $\Delta_x L(x,\nu) = P \succ 0 \rightarrow \min$
* $d(\nu) = -\frac{1}{2}((P^{-1}(A^T\nu + q), (A^T\nu + q)) - (\nu, b)$
* $\nabla_{\nu} d(\nu) = -AP^{-1}A^T\nu - AP^{-1}q - b \rightarrow \nu^* = -A^{-T}q - A^{-T}PA^{-1}b$
* $\Delta_{\nu} d(\nu) = -AP^{-1}A^T \prec 0 \rightarrow \max \ (AP^{-1}A^Tx, x) = (P^{-1}A^Tx, A^Tx) = (P^{-1}y, y) \ge 0 \ \forall y \ (P^{-1} \succ 0)$
* $x_{min} = A^{-1}b$
* Let's rewrite the dual function: $d(\nu) = -\frac{1}{2} (A^T\nu)^TP^{-1}(A^T\nu) - (\nu, APq + b) - (P^{-1}q, q)$
* The dual problem is then as follows: $\max_{\nu} \ -\frac{1}{2} (A^T\nu)^TP^{-1}(A^T\nu) - (\nu, APq + b)$
* Possible implementation of ADMM is as follows: $max_{\alpha, \beta} \ -\frac{1}{2} (A^T\alpha)^TP^{-1}(A^T\alpha) - (\beta, APq + b)$ subject to $\alpha - \beta = 0$
* It's easier for the dual problem because of simpler constraint form.
